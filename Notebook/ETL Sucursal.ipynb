# Accediendo a un Data Lake

### Creaci√≥n de Tablas

En el siguiente ejemplo se muestra como crear una tabla en Azure Databricks

```python
%sql
CREATE TABLE Sucursal
(
Cod_Sucursal INT,Cod_Sucursal_PK INT,Sucursal string ,Latitud string,Longitud string
)
USING CSV
OPTIONS (header 'true', inferSchema 'true', delimiter ',')
LOCATION "/mnt/landing/adventuresales/sales_system/dbAventureWorks/Sucursal/Sucursal.csv"
```

```python
%sql
SELECT * FROM Sucursal
```

```python
%sql
CREATE TABLE DimSucursal
AS
SELECT T.Cod_Sucursal,T.Sucursal,T.Latitud,T.Longitud
FROM Sucursal T
WHERE Cod_Sucursal<>11
```

```python
%sql
SELECT * FROM DimSucursal
```

### Guardado en el Data Lake

En el siguiente ejemplo se muestra como guardar un resultado al Data Lake Store

```python
df = spark.table('DimSucursal')
df.coalesce(1).write.format("com.databricks.spark.csv").mode ("overwrite").option("header", "true").save("/mnt/staging/adventure_sales/sales_system/dbAdventureWorks/DimSucursal")
```

### Borrado de Tablas

En el siguiente ejemplo se muestra como borrar las tablas en un Azure Databricks

```python
%sql
DROP TABLE Sucursal;
DROP TABLE DimSucursal;
```
