# Accediendo a un Data Lake

### Creaci√≥n de Tablas

En el siguiente ejemplo se muestra como crear una tabla en Azure Databricks

```python
%sql
CREATE TABLE Ventas
(
Cod_Producto STRING,
Cod_Cliente STRING,
Cod_Sucursal STRING,
NumeroOrden STRING,
Cantidad STRING,
PrecioUnitario STRING,
CostoUnitario STRING,
Impuesto STRING,
Flete STRING,
FechaOrden STRING,
FechaEnvio STRING,
FechaVencimiento STRING,
Cod_Promocion STRING
)
USING CSV
OPTIONS (header 'true', inferSchema 'true', delimiter ',')
LOCATION "/mnt/landing/adventuresales/sales_system/dbAventureWorks/Ventas/VentasInternet.csv"
```

```python
%sql
SELECT * FROM Ventas
```

```python
%sql
CREATE TABLE FactVentas
AS
SELECT T.Cod_Sucursal,T.Cod_Producto,T.Cantidad,T.FechaOrdeN
FROM Ventas T
```

```python
%sql
SELECT * FROM FactVentas
```

### Guardado en el Data Lake

En el siguiente ejemplo se muestra como guardar un resultado al Data Lake Store

```python
df = spark.table('FactVentas')
df.coalesce(1).write.format("com.databricks.spark.csv").mode ("overwrite").option("header", "true").save("/mnt/staging/adventure_sales/sales_system/dbAdventureWorks/FactVentas")
```

### Borrado de Tablas

En el siguiente ejemplo se muestra como borrar las tablas en un Azure Databricks

```python
%sql
DROP TABLE Ventas;
DROP TABLE FactVentas;
```
