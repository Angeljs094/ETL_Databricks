# Accediendo a un Data Lake

## Montado

En el siguiente script realizamos el montado de nuestro Data Lake hacia Azure Databricks

```python
configs = {"fs.azure.account.auth.type": "OAuth",
           "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
           "fs.azure.account.oauth2.client.id": "<ClienteID>",
           "fs.azure.account.oauth2.client.secret" : "<SecretID>",
           "fs.azure.account.oauth2.client.endpoint": "https://login.microsoftonline.com/<TenanID>/oauth2/token"}
dbutils.fs.mount(
  source = "abfss://source@stacounttangel.dfs.core.windows.net/",
  mount_point = "/mnt/source20240810",
  extra_configs = configs)
```

```python
configs = {"fs.azure.account.auth.type": "OAuth",
           "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
           "fs.azure.account.oauth2.client.id": "<ClienteID>",
           "fs.azure.account.oauth2.client.secret" : "<SecretID>",
           "fs.azure.account.oauth2.client.endpoint": "https://login.microsoftonline.com/<TenanID>/oauth2/token"}
dbutils.fs.mount(
  source = "abfss://raw@stacounttangel.dfs.core.windows.net/",
  mount_point = "/mnt/raw20240810",
  extra_configs = configs)
```

### Desmontado

Con el siguiente ejemplo se muestra como realizar un desmontado

```python
dbutils.fs.unmount("/mnt/raw")
dbutils.fs.unmount("/mnt/source")
```
